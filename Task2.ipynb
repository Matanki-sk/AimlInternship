```python
import pandas as pd

# Load the dataset from the uploaded CSV
df = pd.read_csv('Titanic-Dataset.csv')

# Show first 5 rows
df.head()
```

```python
# Info about data types and missing values
df.info()
```

```python
df.describe(include='all')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Age'])
plt.title('Boxplot of Passenger Ages')
plt.xlabel('Age')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include=['float64', 'int64']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix (Numeric Features Only)')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Sex')
plt.ylabel('Survival Probability')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Probability')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Embarked', y='Survived', data=df)
plt.title('Survival Rate by Embarked Port')
plt.ylabel('Survival Probability')
plt.show()
```

```python
print("""
Observations from EDA:

- Females had a higher survival rate than males.
- Passengers in first class (Pclass=1) had better survival chances.
- The age distribution is right-skewed with many young passengers.
- Missing values exist in 'Age' and 'Cabin' columns.
- Fare tends to be higher for survivors.
- Embarked location shows differences in survival rates.
""")
```
